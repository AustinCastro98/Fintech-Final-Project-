As you can see, our expenses are not inline with the 20/30/50 budgeting rule at all. Lets create a bank simulator that will help us keep our expenses inline. 

Below we will use classes and methods to create a bank account. 

```python
class Checkings:
    def __init__(self,balance=0):
        self.balance = balance
    def __str__(self):
        return f'Account balance: ${self.balance}'

    def deposit(self,dep_amt):
        self.balance += dep_amt
        print(f'Your deposit for ${dep_amt} was accepted.\nYour current checking balance is ${self.balance}.')

    def withdraw(self,wd_amt):
        if self.balance >= wd_amt and wd_amt<=500:
            self.balance -= wd_amt
            print(f'Your withdrawal for ${wd_amt} was accepted.\nYour current checking balance is ${self.balance}.')
        elif wd_amt>500:
            print('Withdrawal amount cannot exceed $500')
        else:
            print('Funds Unavailable!')
```
